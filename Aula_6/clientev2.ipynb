```python
import pandas
```

```python
clientes_v2 = {'nomes':['Ana', 'João', 'Fernando', 'Silva', 'William', 'Beatriz', 'Cássia', 'Maria'],
			'idades':[54, 48, 31, 23, 34, 24, 40, 35],
			'salario':[1220, 5020, 1500, 1890, 7000, 2000, 1220, 1220]}
```

```python
df = pandas.DataFrame(clientes_v2)
```

```python
# media de idade
print('A media de idades é de:', df['idades'].mean())
```

```python
# media de salario
print('A media de Salarios é de: R$', df['salario'].mean())
```

```python
# Qual o mais jovem / velho nome?
print(df, '\n')
mais_jovem = df.sort_values(['idades'], ascending = True)
mais_velho = df.sort_values(['idades'], ascending = False)


print('A mais jovem é', mais_jovem.iloc[0]['nomes'], 
      ' e o mais velho é ', mais_velho.iloc[0]['nomes'])

```

```python
# Quem ganha o maior salario, nome e idade?

maior_salario = df.sort_values(['salario'], ascending = False)
print(maior_salario, '\n')
ganha_mais = maior_salario.iloc[0]
print(ganha_mais['nomes'], ganha_mais['idades'])
```

```python
df.to_csv ('cadastro.csv', index = False)
```
